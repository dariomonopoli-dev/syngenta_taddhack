```python
import pandas as pd
from api.CE_Hub_api import get_data

data = get_data()

df = pd.DataFrame(data)


df_pivot = df.pivot(
    index="date", columns="measureLabel", values="value"
).reset_index()


df_pivot.columns = [
    col.replace(" ", "_").replace("(", "").replace(")", "").replace("/", "_")
    for col in df_pivot.columns
]


df_pivot
```
