```python
import pandas as pd
from api.CE_Hub_api import get_data

data = get_data()

df = pd.DataFrame(data)


df_pivot = df.pivot(
    index="date", columns="measureLabel", values="dailyValue"
).reset_index()


df_pivot.columns = [
    col.replace(" ", "_").replace("(", "").replace(")", "").replace("/", "_")
    for col in df_pivot.columns
]


df_pivot["TempAir_DailyMin_C"] = df_pivot["TempAir_DailyMin_C"].astype(float)
df_pivot["TempAir_DailyMax_C"] = df_pivot["TempAir_DailyMax_C"].astype(float)

df_pivot
```
